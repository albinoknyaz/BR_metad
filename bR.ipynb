```python
#Rgyr calculation
protein = u.select_atoms("protein") 
for ts in u.trajectory:
        print("Frame: {0:5d}, Time: {1:8.3f} ps".format(ts.frame, u.trajectory.time))
        print("Rgyr: {0:g} A".format(protein.atoms.radius_of_gyration()))
        
Rgyr = []
for ts in u.trajectory:
   Rgyr.append((u.trajectory.time, protein.radius_of_gyration()))
Rgyr = np.array(Rgyr)
```

```python
#Rgyr plot
import matplotlib.pyplot as plt

plt.plot(Rgyr[:,0], Rgyr[:,1], 'k', lw=0.2)
plt.xlabel('time (ps)')
plt.ylabel('radius of gyration $R_G$ ($\AA$)')
plt.savefig('Rgyr.pdf')
plt.show()
```

```python
w = MDAnalysis.Universe('/home/knyazeva/bR/Structures/charmm-gui/gromacs/mead.pdb')
def dipole_moment(u):
    C = u.select_atoms("protein").center_of_mass()
    r = u.select_atoms('protein').positions
    q = w.select_atoms('protein').occupancies
    return (np.sum(q[np.newaxis].T*(r-C),0))
k=dipole_moment(w)
k
```

```python
#dipole moment calculation
k=u.trajectory[0]
dip=dipole_moment(u) 
norm=np.linalg.norm(dip)
dipole=[]
for k in u.trajectory:
    print("Frame: {0:5d}, Time: {1:8.3f} ps".format(k.frame, u.trajectory.time))
    print('Dipole moment: {0:8.3f} {1:8.3f} {2:8.3f} '.format(dip[0], dip[1], dip[2]))
    print('Norm {0:8.3f} '.format(norm))    
    dipole.append((u.trajectory.time, dip[0], dip[1], dip[2], norm))
dipole = np.array(dipole)
```

```python
import MDAnalysis
import MDAnalysis.analysis.hbonds
import numpy as np
u = MDAnalysis.Universe('/home/knyazeva/bR/Structures/charmm-gui/gromacs/step7_1.gro', 
                        '/home/knyazeva/bR/Structures/charmm-gui/gromacs/step7_1.trr')

path = u.select_atoms("(resid 96) or (resid 216) or (resid 85) or (resid 82) or (resid 204) or (resid 194) ")

for ts in u.trajectory:
    zch = u.select_atoms("protein and (resid 167 or resid 191)").positions
    zh=float(max(np.array(zch[0:,2:])))
    zl=float(min(np.array(zch[0:,2:])))
    water = u.select_atoms("resname TIP3 and (prop z <=%f or prop z >=%f)"%(zh,zl))
    ch = path | water
    h = MDAnalysis.analysis.hbonds.HydrogenBondAnalysis (u, selection1="ch", selection2="ch", distance=3.0, angle=120.0)
    h.run()
    
    h_presence = []
    for frame in h.timeseries:
        if frame:
            h_presence.append(True)
        else:
            h_presence.append(False)
    p_h = float(sum(h_presence))/len(h_presence)
    print("Fraction of time with Hydrogen Bonds present: {}".format(p_h))
```
