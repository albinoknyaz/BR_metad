```python
#function of dipole moment
w = MDAnalysis.Universe('/home/knyazeva/bR/Structures/charmm-gui/gromacs/mead.pdb')
def dipole_moment(u):
    C = u.select_atoms("protein").center_of_mass()
    r = u.select_atoms('protein').positions
    q = w.select_atoms('protein').occupancies
    return (np.sum(q[np.newaxis].T*(r-C),0))
k=dipole_moment(w)
k
```

```python
#dipole moment calculation
k=u.trajectory[0]
dip=dipole_moment(u) 
norm=np.linalg.norm(dip)
dipole=[]
for k in u.trajectory:
    print("Frame: {0:5d}, Time: {1:8.3f} ps".format(k.frame, u.trajectory.time))
    print('Dipole moment: {0:8.3f} {1:8.3f} {2:8.3f} '.format(dip[0], dip[1], dip[2]))
    print('Norm {0:8.3f} '.format(norm))    
    dipole.append((u.trajectory.time, dip[0], dip[1], dip[2], norm))
dipole = np.array(dipole)
```

```python
#HBounds in channel


for ts in trj.trajectory:
    zch = trj.select_atoms("protein and (resid 96 or resid 204 or resid 194)").positions
    C = trj.select_atoms("protein").center_of_geometry()
    Zh=max(np.array(zch[0:,2:]))-C-3
    Zl=min(np.array(zch[0:,2:]))-C-3

zh=np.average(Zh)
zl=np.average(Zl)
    
selection_string="((resid 96) or (resid 216) or (resid 85) or (resid 82) or (resid 204) or (resid 194)) or  "
selection_string+="(resname TIP3 and cyzone 11.0 %f %f protein)"%(zh,zl)

h_channel = MDAnalysis.analysis.hbonds.HydrogenBondAnalysis (trj, selection1=selection_string,
                                                selection2=selection_string,
                                                distance=3.0, angle=120.0,
                                                update_selection1=True, update_selection2=True)
h_channel.run()
    
h_presence = []
for frame in h_channel.timeseries:
    h_presence.append(len(frame))
p_h = float(np.average(h_presence))
print("Fraction of time with Hydrogen Bonds present: {}".format(p_h))
```
